#Analisando as informação com gráficos


```python
#organizar as informações em ordem decrescente para uma melhor visualização
vendas_decrescente = vendas.sort_values('Quantidade de Vendas', ascending=False)
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(16,6))
sns.barplot(data= vendas_decrescente, x = 'Quantidade de Vendas', y = 'Nome do Curso', orient ='h',palette='magma')
plt.title('Número de vendas por curso', fontsize=22)
plt.xlabel('Quantidade de vendas', fontsize=16)
plt.ylabel('Nome do curso', fontsize=16)

#removendo as bordas do gráfico
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.show()
```

# Visualizar a distribuição das vendas ao longo do tempo através de gráficos.

```python
sns.set_theme(style = 'whitegrid')
plt.figure(figsize=(16,6))
sns.lineplot(vendas, x = 'Data', y = 'Quantidade de Vendas', color = 'black', marker = 'o')
plt.title('Distribuição das vendas no mês de janeiro')
plt.xticks(rotation=70)
plt.show()
```

Outra forma de visualizar é com o uso do gráfico de barras. Nesse caso, as datas serão organizar de forma decrescente do dia com maior número de vendas para o dia com o menor número de vendas

```python
#organizar o DataFrame em ordem decrescente
vendas_ordenada = vendas.sort_values('Quantidade de Vendas', ascending=False)

#selecionar o tema do gráfico e alterar o tamanho da imagem
sns.set_theme(style='white')
plt.figure(figsize=(16,6))

#criar o gráfico de barras
sns.barplot(data=vendas_ordenada, x='Data', y='Quantidade de Vendas', palette = 'magma')

#nomear os eixos e adicionar o título
plt.title('Distribuição das vendas no mês de janeiro')
plt.xlabel('Datas')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=70)
plt.show()
```

#Explorando as relações entre a quantidade de cursos vendidos e o preço

```python
#Visualização da relação com um gráfico de dispersão
sns.set_theme(style = 'white')
plt.figure(figsize=(12,5))
sns.scatterplot(data = vendas, x='Quantidade de Vendas', y = 'Preço Unitário', hue ='Quantidade de Vendas', palette='flare', s=100)
plt.title('Relação entre a quantidade de cursos vendidos e o preço unitário')
plt.show()
```
